### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Error handling with exceptions
if not isinstance(labels, list) or not isinstance(sizes, list):
    raise TypeError("Labels and sizes must be lists")
if len(labels) != len(sizes):
    raise ValueError("Labels and sizes must have the same length")
if not all(isinstance(s, (int, float)) and s >= 0 for s in sizes):
    raise ValueError("Sizes must be non-negative numbers")
if sum(sizes) == 0:
    raise ValueError("Sizes must contain at least one positive value")

# Create pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Customization
plt.title('Programming Language Popularity')

# Show plot
plt.show()


```
