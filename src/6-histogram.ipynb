### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Enhanced error handling
assert isinstance(data, list), "Data must be a list"
assert len(data) > 0, "Data list cannot be empty"
assert all(isinstance(i, (int, float)) for i in data), "All elements must be numeric"
assert all(i >= 0 for i in data), "Data must contain non-negative values"
assert any(i > 0 for i in data), "Data must contain at least one positive value"

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=5, color='blue', edgecolor='black')

# Customization
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()

```
