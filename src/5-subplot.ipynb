### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np
# Data for all plots
x_line = np.arange(0, 11)
y_line = x_line ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line chart
axs[0, 0].plot(x_line, y_line, marker='o', color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('x squared')
axs[0, 0].grid(True)

# Bar chart
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), 
                       textcoords="offset points", xytext=(5,5), ha='center')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

# Pie chart
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
